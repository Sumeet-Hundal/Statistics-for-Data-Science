```python
m1 = sm.ols('y ~ x', data = xd1)
m1 = m1.fit()
m1.summary()
```

```python
intercept, slope = m1.params
xd1['y_est'] = intercept + slope * xd1.x
```

```python
plt.plot(xd1.x, xd1.y, 'o', label = 'Observations')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Set 1')

# Draw estimates
plt.plot(xd1.x, xd1.y_est, 'r+', label='Estimates')

# Plot regression line
plt.plot(xd1.x, xd1.y_est, 'r-', label='Regression line')

# Plot errors
plt.vlines(xd1.x, xd1.y_est, xd1.y, linestyles='dashed', label = 'errors')

# Add legend
plt.legend()
plt.show()
```

### [Set 1] The is no need to transfrom the data as can been seen from the plot with errors. The error from regression line is small. Also the r^sq= 97.9% which is very high and std error=0.074 is very low.

### [Set 3] There is a need for transformation as the data is not linear. The r^sq=75% which is still good and the plot shows that regression does pass through much of the data but it can be improved with transformation.

```python
ml3 = sm.ols(formula='xdl3.y1 ~ xdl3.x1', data = xdl3)
ml3 = ml3.fit()
ml3.params
intercept, slope = ml3.params
xdl3['y1_est'] = intercept + slope * xdl3.x1
```

```python
plt.plot(xdl3.x1, xdl3.y1, 'o', label = 'Observations')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('Set 3 Log(x&y)')

# Draw estimates
plt.plot(xdl3.x1, xdl3.y1_est, 'r+', label='Estimates')

# Plot regression line
plt.plot(xdl3.x1, xdl3.y1_est, 'r-', label='Regression line')

# Plot errors
plt.vlines(xdl3.x1, xdl3.y1_est, xdl3.y1, linestyles='dashed', label = 'errors')

# Add legend
plt.legend()
plt.show()
```
